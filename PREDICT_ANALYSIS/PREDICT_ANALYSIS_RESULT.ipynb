```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Filter the DataFrame to exclude rows with any value greater than 100000 in specified columns
filtered_df = daily_df[(daily_df['Bagging_LR_Percentage_Error'] <= 20) &
                          (daily_df['Bagging_GB_Percentage_Error'] <= 20)]

# Set up the matplotlib figure
plt.figure(figsize=(14, 8))

# Create a scatter plot for each percentage error column
sns.scatterplot(data=filtered_df, x=filtered_df.index, y='Bagging_LR_Percentage_Error', label='Bagging_LR_Percentage_Error', color='blue')
sns.scatterplot(data=filtered_df, x=filtered_df.index, y='Bagging_GB_Percentage_Error', label='Bagging_GB_Percentage_Error', color='red')

# Set plot labels and title
plt.xlabel('Index')
plt.ylabel('Percentage Error')
plt.title('Variation of Percentage Error with respect to Index')

# Show plot
plt.legend()
plt.show()
```

```python
# Filter the DataFrame to exclude rows with any value greater than 100000 in any of the three columns
filtered_df = daily_df[(daily_df['period'] == 'daily') &
                 ((daily_df[['Bagging_LR_Percentage_Error', 'Bagging_GB_Percentage_Error']] <= 20).all(axis=1))]

# Determine the lowest percentage error value among the three columns
filtered_df['lowest_error'] = filtered_df[['Bagging_LR_Percentage_Error', 'Bagging_GB_Percentage_Error']].min(axis=1)

# Set up the matplotlib figure
plt.figure(figsize=(14, 8))

# Create a scatter plot for the lowest percentage error column
sns.scatterplot(data=filtered_df, x=filtered_df.index, y='lowest_error', color='blue')

# Set plot labels and title
plt.xlabel('Index')
plt.ylabel('Lowest Percentage Error')
plt.title('Variation of Lowest Percentage Error with respect to Index')

# Show plot
plt.show()
```
